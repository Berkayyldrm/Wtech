**2. Hava durumu modeli**

Bu alıştırmada, önceki dersteki sıcaklık verileriyle çalışacaksınız. Bu ödevi tamamlamak için, gist dosyasına, aşağıdaki görevlere ilişkin çözümlerinizi içeren Jupyter notebook'u bir link olarak ekleyin:

```python
import pandas as pd
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

* Hedef değişkeninizin görünür sıcaklık ve sıcaklık arasındaki fark olduğu doğrusal bir regresyon modeli oluşturun. Açıklayıcı değişkenler olarak nem ve rüzgar hızı kullanın. Şimdi, modelinizi OLS kullanarak tahmin edin. Tahmin edilen katsayılar istatistiksel olarak anlamlı mıdır? Tahmini katsayılar önceki beklentileriniz doğrultusunda mı? Tahmin edilen katsayıları yorumlayınız. Hedef ve açıklayıcı değişkenler arasındaki ilişkiler nelerdir?

```python
import statsmodels.api as sm
weather_df["Yeni"] = weather_df["Apparent Temperature (C)"] - weather_df["Temperature (C)"]
Y = weather_df["Yeni"]
X = weather_df[["Humidity", "Wind Speed (km/h)"]]
X = sm.add_constant(X)
```

```python
sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

Tahmin edilen katsayılar istatiksel olarak anlamlıdır(P değerleri 0 a yakındır). Nem ve rüzgar hızı, görünen sıcaklık ile sıcaklık farkıyla ters orantılıdır.

* Ardından, yukarıdaki modele nem ve rüzgar hızı etkileşimini dahil edin ve OLS'yi kullanarak modeli tahmin edin. Katsayılar istatistiksel olarak anlamlı mıdır? Nem ve rüzgar hızı için tahmini katsayıların işaretleri değişti mi? Tahmin edilen katsayıları yorumlayınız.

```python
Y = weather_df["Yeni"]
weather_df['hum_wind'] = weather_df.Humidity * weather_df["Wind Speed (km/h)"]
X = weather_df[["Humidity", "Wind Speed (km/h)","hum_wind"]]

X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

Tahmin edilen katsayılar istatiksel olarak anlamlıdır(P değerleri 0 a yakındır).
Katsayıların işaretleri değişmiştir. Yeni değişken tüm katsayıları etkilemiştir.  Nem ve rüzgar hızı, görünen sıcaklık ile sıcaklık farkıyla doğru orantılıdır. hum_wind, görünen sıcaklık ile sıcaklık farkıyla ters orantılıdır.

**3.Ev fiyatları modeli**

Bu alıştırmada, ev fiyatları modelinizi yorumlayacaksınız. Bu ödevi tamamlamak için, gist dosyasına, aşağıdaki görevlere ilişkin çözümlerinizi içeren Jupyter notebook'u bir link olarak ekleyin:

```python
house_prices_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
house_prices_df.head()
```

* Ev fiyatları modelinizi tekrar çalıştırın ve sonuçları yorumlayın. Hangi özellikler istatistiksel olarak anlamlı ve hangileri değildir?

```python
house_prices_df = house_prices_df.dropna(axis=1, how="any")
Y = house_prices_df["SalePrice"]
X = house_prices_df[["OverallQual", "TotalBsmtSF", "GrLivArea", "GarageCars", "GarageArea"]]
X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

GarageArea değişkeninin P değeri biraz yüksek olduğundan istatiksel olarak anlamsız sayılabilir. Diğerlerinin P değerleri 0 dır. Bunlar anlamlı sayılabilirler.

* Şimdi, anlamsız özellikleri modelinizden hariç tutun. Bir şey değişti mi?

```python
Y = house_prices_df["SalePrice"]
X = house_prices_df[["OverallQual", "TotalBsmtSF", "GrLivArea", "GarageCars"]]
X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

Tüm P değerleri 0 çıktı. Anlamsız değişken kalmadı.

* Ev fiyatları ile olan ilişkilerini belirleyerek istatistiksel olarak anlamlı katsayıları yorumlayın. Ev fiyatları üzerinde hangi özellikler daha belirgin bir etkiye sahiptir?

GarageCars, GrLivArea, TotalBsmt, OverallQual değişkenleri pozitif yüksek katsayılara sahiptir. GarageCars ve OverallQual daha belirgin etkiye sahiptir.
